# PARTE III Muertes Totales en Chile

```python
Muertes_Totales = dataframe2
Muertes_Totales = Muertes_Totales.reset_index()
Muertes_Totales = Muertes_Totales.drop(['index'], axis = 1)
MUERTES_TOTALES = Muertes_Totales[(Muertes_Totales['ANO_DEF']== 2021) | (Muertes_Totales['ANO_DEF'] ==2020)]

MUERTES_TOTALES= MUERTES_TOTALES.groupby(['FECHA_DEF']).size().reset_index(name = 'NUMERO_DEF')

MUERTES_TOTALES
```

```python
MUERTES_TOTALES['FECHA_DEF'] =pd.to_datetime(MUERTES_TOTALES['FECHA_DEF'])

MUERTES_TOTALES=MUERTES_TOTALES.query("FECHA_DEF >= '2020-03-16' and FECHA_DEF <= '2021-04-20'")

MUERTES_TOTALES = MUERTES_TOTALES.reset_index()

MUERTES_TOTALES = MUERTES_TOTALES.drop(['index'],axis=1)

MUERTES_TOTALES['FECHA_DEF']=pd.to_datetime(MUERTES_TOTALES['FECHA_DEF'])

MUERTES_TOTALES = MUERTES_TOTALES.resample('W', on ='FECHA_DEF').sum()

MUERTES_TOTALES = MUERTES_TOTALES.reset_index()

```

```python
XXX=MUERTES_TOTALES['FECHA_DEF']
YYY=MUERTES_TOTALES['NUMERO_DEF']
fig, ax = plt.subplots()
ax.plot(XXX,YYY)
plt.title("Defunciones Totales en Chile")
plt.show
```
