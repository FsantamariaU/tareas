```python
melted1 = df.melt(id_vars=['Region','Codigo region','Comuna','Codigo comuna','Poblacion'],var_name='Edad')
```

```python
melted1.Edad = melted1.Edad.astype(int)
```

```python
def sumar(X):
    if X<65:
        return 'Edad<65'
    else: 
        return 'Edad65ymas'
```

```python
melted1['Range']=melted1.Edad.apply(sumar)

melted1
```

```python
melted1 = melted1.pivot_table(index =['Region','Codigo region','Comuna','Codigo comuna','Poblacion'], columns = 'Range' , values = 'value').reset_index()
```

```python
melted1.head()
```

```python
melted1['Prop65'] = (melted1['Edad65ymas'] / (melted1['Edad<65'] + melted1['Edad65ymas']))
melted1.head()
```
